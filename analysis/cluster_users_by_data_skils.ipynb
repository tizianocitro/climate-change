```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# users data from results
users = np.array([
    [4, 4], # user1
    [3, 3], # user2
    [4, 4], # user3
    [4, 4], # user4
    [2, 3], # user5
    [3, 2], # user6
    [3, 4], # user7
    [3, 3], # user8
    [1, 1], # user9
    [5, 5], # user10
    [4, 3], # user11
    [1, 2], # user12
    [4, 4], # user13
    [1, 1], # user14
    [4, 4], # user15
    [3, 4], # user16
    [3, 2], # user18
    [3, 3], # user19
    [2, 4], # user20
    [3, 2], # user21
    [2, 4], # user22
    [4, 1], # user23
])
```

```python
# create a KMeans instance with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=0, n_init=10)

# predict the cluster for each user
clusters = kmeans.fit_predict(users)
```

```python
# print the cluster centers
print("Cluster centers:\n", kmeans.cluster_centers_)

# print the cluster assignments
print("Cluster assignments:", clusters)
```

```python
# count the number of points in each cluster
counts = np.bincount(clusters)
for i, count in enumerate(counts):
    print(f"cluster {i}: {count} elements")
```

```python
# determine unique user coordinates
unique_users, unique_counts = np.unique(users, axis=0, return_counts=True)
for user, count in zip(unique_users, unique_counts):
  print(f'unique point: {user} has {count} users')
print()

plt.figure(figsize=(10, 8))
plt.scatter(users[:, 0], users[:, 1], c=clusters, cmap='viridis', label='Users')
plt.xlabel('Data Manipulation')
plt.ylabel('Data Visualization')
plt.title('Users Clustered By Data Manipulation & Visualization Skills')

# annotate points with overlapping users
for (x, y), count in zip(unique_users, unique_counts):
    users_at_point = [f'user{i + 1}' for i, user in enumerate(users) if (user == [x, y]).all()]
    label_text = '\n'.join(users_at_point)
    plt.annotate(label_text, (x, y), textcoords="offset points", xytext=(0, 10), ha='center')

plt.legend()
plt.show()
```
